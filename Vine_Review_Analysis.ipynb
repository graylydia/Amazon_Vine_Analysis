```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table2.csv")
vine_df
```

```python
total_df = vine_df.loc[vine_df['total_votes'] >= 20]
total_df
```

```python
filtered_df = total_df.loc[total_df['helpful_votes'] / total_df['total_votes'] >= 0.5]
filtered_df
```

```python
paid_reviews_df = filtered_df.loc[filtered_df['vine'] == 'Y']
paid_reviews_df
```

```python
unpaid_reviews_df = filtered_df.loc[filtered_df['vine'] == 'N']
unpaid_reviews_df
```

```python
paid_total_reviews = len(paid_reviews_df)
paid_total_reviews
```

```python
unpaid_total_reviews = len(unpaid_reviews_df)
unpaid_total_reviews
```

```python
paid_five_star = len(paid_reviews_df.loc[paid_reviews_df['star_rating'] == 5.0])
paid_five_star
```

```python
unpaid_five_star = len(unpaid_reviews_df.loc[unpaid_reviews_df['star_rating'] == 5.0])
unpaid_five_star
```

```python
paid_percentage = (paid_five_star / paid_total_reviews) * 100
paid_percentage
```

```python
unpaid_percentage = (unpaid_five_star / unpaid_total_reviews) * 100
unpaid_percentage
```
